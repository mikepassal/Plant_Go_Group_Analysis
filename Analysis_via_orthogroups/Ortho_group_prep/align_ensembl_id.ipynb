```python
import pandas as pd 
import CococoNet_reader 
import gc 
import numpy as np
```

```python
gene_xref = pd.read_csv('/data/passala/OrthoDB_data/odb11v0_gene_xrefs.tab',names = ['Orthodb Gene','External ID','Database Name'],sep = '\t')
```

```python
species_with_nets = pd.read_csv('/data/passala/Generated_Tables/Reference_tables/Species_name_resolver.csv')
species_with_nets = species_with_nets[:18]
# # species_with_nets = species_with_nets.drop(index = [13,14])
# # species_with_nets.loc[0,'Taxa ID'] = 39947 
taxa_to_keep = species_with_nets['Taxa ID'].astype(str).to_list()
taxa_to_keep_int = species_with_nets['Taxa ID'].to_list()
taxa_to_keep
```

```python
gene_subset = gene_xref.loc[gene_xref['Database Name']=='NCBIgid']
```

```python
gene_subset['Species']  = gene_subset['Orthodb Gene'].str.split('_').str[0]
```

```python
gene_subset
```

```python
species_subset = gene_subset.loc[gene_subset['Species'].isin(taxa_to_keep)]
```

```python
del gene_subset
del gene_xref
gc.collect()
```

```python
rice_gene_info = pd.read_csv('/data/CoCoCoNet/geneInfo/rice_info.csv')
rice_gene_info
```

```python
arab_gene_info = pd.read_csv('/data/CoCoCoNet/geneInfo/arabidopsis_info.csv')
soybean_gene_info = pd.read_csv('/data/CoCoCoNet/geneInfo/soybean_info.csv')
maize_gene_info =pd.read_csv('/data/CoCoCoNet/geneInfo/maize_info.csv')
```

```python
merged_ncbi_and_species = merged_ncbi_and_species.merge(how = 'left', right = rice_gene_info[['EntrezID','NetworkIDs']],left_on='GeneID',right_on='EntrezID')
merged_ncbi_and_species['NetworkIDs'] = merged_ncbi_and_species['NetworkIDs'].fillna(value = merged_ncbi_and_species['Symbol'])
merged_ncbi_and_species = merged_ncbi_and_species.drop(columns = ['Symbol','EntrezID'])
merged_ncbi_and_species = merged_ncbi_and_species.rename(columns = {'NetworkIDs':'Symbol'})
merged_ncbi_and_species
```

```python
merged_ncbi_and_species = merged_ncbi_and_species.merge(how = 'left', right = arab_gene_info[['EntrezID','NetworkIDs']],left_on='GeneID',right_on='EntrezID')
merged_ncbi_and_species['NetworkIDs'] = merged_ncbi_and_species['NetworkIDs'].fillna(value = merged_ncbi_and_species['Symbol'])
merged_ncbi_and_species = merged_ncbi_and_species.drop(columns = ['Symbol','EntrezID'])
merged_ncbi_and_species = merged_ncbi_and_species.rename(columns = {'NetworkIDs':'Symbol'})
merged_ncbi_and_species
```

```python
merged_ncbi_and_species = merged_ncbi_and_species.merge(how = 'left', right = soybean_gene_info[['EntrezID','NetworkIDs']],left_on='GeneID',right_on='EntrezID')
merged_ncbi_and_species['NetworkIDs'] = merged_ncbi_and_species['NetworkIDs'].fillna(value = merged_ncbi_and_species['Symbol'])
merged_ncbi_and_species = merged_ncbi_and_species.drop(columns = ['Symbol','EntrezID'])
merged_ncbi_and_species = merged_ncbi_and_species.rename(columns = {'NetworkIDs':'Symbol'})
merged_ncbi_and_species
```

```python
merged_ncbi_and_species = merged_ncbi_and_species.merge(how = 'left', right = maize_gene_info[['EntrezID','NetworkIDs']],left_on='GeneID',right_on='EntrezID')
merged_ncbi_and_species['NetworkIDs'] = merged_ncbi_and_species['NetworkIDs'].fillna(value = merged_ncbi_and_species['Symbol'])
merged_ncbi_and_species = merged_ncbi_and_species.drop(columns = ['Symbol','EntrezID'])
merged_ncbi_and_species = merged_ncbi_and_species.rename(columns = {'NetworkIDs':'Symbol'})
merged_ncbi_and_species
```
