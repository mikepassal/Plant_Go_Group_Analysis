```python
import pyreadr 
import pandas as pd
import pickle 
```

```python
og2genes = pd.read_csv('/data/passala/OrthoDB_data/odb11v0_OG2genes.tab',sep = '\t', names=['Orthogroup','Gene'])
og2genes
```

```python
og2genes['Taxa ID'] = og2genes['Gene'].str.replace('_.*','',regex = True)
og2genes
```

```python
og2genes.to_csv('/data/passala/OrthoDB_data/cleaned_og2genes.csv',index=False)
file_og2genes = open('/data/passala/Generated_Tables/Comparing_all_go_groups_across_species/OrthoDB_python_files/orthodb_v11_og2genes.p','wb')
pickle.dump(og2genes,file_og2genes)
file_og2genes.close()
```
