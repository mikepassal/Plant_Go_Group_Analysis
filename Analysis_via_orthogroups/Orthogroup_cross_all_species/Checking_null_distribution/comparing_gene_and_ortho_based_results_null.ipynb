```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib as mpl
mpl.rcParams['pdf.fonttype'] = 42
mpl.rcParams['ps.fonttype'] = 42
mpl.rcParams['svg.fonttype'] = 'none'
```

```python
nullwise = pd.read_csv('/data/passala/Post_dump_files/Generated_Tables/Comparing_all_orthogorups_across_plants/Ortho_cross_all_tables/egad_and_fc_over_2_species_orthowise_null.csv',index_col= 0 )
nullwise['Conservation Between'] = 'Randomly Drawn Orthogroups'
nullwise
```

```python
orthowise = pd.read_csv('/data/passala/Generated_Tables/Comparing_all_orthogorups_across_plants/Ortho_cross_all_tables/egad_and_fc_over_2_species_orthowise.csv', index_col= 0 )
orthowise['Conservation Between'] = 'Orthogroups'
orthowise
```

```python
merged_df = pd.concat([nullwise,orthowise])
merged_df
```

```python
fig, ax = plt.subplots(figsize = (25,8))
fig.patch.set_facecolor('white')
ax = sns.violinplot(data = merged_df, x = 'EGAD Species Present', y = 'AVG FC', hue = 'Conservation Between', split = False, density_norm = 'area', inner = 'quartile', width=1)

plt.ylabel("Conservation of Coexpression", fontsize = 20)
plt.xlabel("Number of studied species with orthogroup", fontsize = 20)
sns.despine()

ax.grid(False)
ax.spines["bottom"].set_color("black")
ax.spines["left"].set_color("black")
sns.move_legend(ax, 'lower right')
#plt.xlim((-7,7))
# plt.ylim((0, 1250))
#plt.title("Conservation of orthogroup coexpression captures conserved  \n gene relationships missed in genewise comparisons", fontsize = 20)
plt.xticks(fontsize = 15,)
plt.yticks(fontsize = 15)
plt.savefig('/data/passala/Post_dump_files/Plots/orthogroup_improvement_over_pairwise.pdf', transparent = True)
```
