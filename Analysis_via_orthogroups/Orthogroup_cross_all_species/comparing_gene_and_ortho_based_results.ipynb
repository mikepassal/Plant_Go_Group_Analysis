```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib as mpl
mpl.rcParams['pdf.fonttype'] = 42
mpl.rcParams['ps.fonttype'] = 42
mpl.rcParams['svg.fonttype'] = 'none'
```

```python
genewise = pd.read_csv('/data/passala/Generated_Tables/Comparing_all_orthogorups_across_plants/Ortho_cross_all_tables/egad_and_fc_over_2_species_genewise.csv',index_col= 0 )
genewise['Conservation Between'] = 'Genes'
genewise
```

```python
orthowise = pd.read_csv('/data/passala/Generated_Tables/Comparing_all_orthogorups_across_plants/Ortho_cross_all_tables/egad_and_fc_over_2_species_orthowise.csv', index_col= 0 )
orthowise['Conservation Between'] = 'Orthogroups'
orthowise
```

```python
merged_df = pd.concat([genewise,orthowise])
merged_df
```

```python
orthowise = orthowise.rename(columns= {'AVG FC':'AVG FC Ortho','FC Species Present':'FC Species Present Ortho','Conservation Between':'Conservation Between Ortho'})
```

```python
orthowise
```

```python
orthowise[['AVG FC Ortho','FC Species Present Ortho','Conservation Between Ortho']]
```

```python
merged_wide = genewise.merge(how = 'inner',right= orthowise[['AVG FC Ortho','FC Species Present Ortho','Conservation Between Ortho']], right_index= True, left_index= True)
```

```python
merged_wide
```

```python
merged_wide['Gene minus Ortho'] = merged_wide['AVG FC'] - merged_wide['AVG FC Ortho']
merged_wide
```

```python
merged_wide.sort_values(by = 'Gene minus Ortho').head(177)
```

```python
fig, ax = plt.subplots(figsize = (15,8))
fig.patch.set_facecolor('white')
ax = sns.violinplot(data = merged_df, x = 'EGAD Species Present', y = 'AVG FC', hue = 'Conservation Between', split = True, inner = 'quartile')

plt.ylabel("Conservation of Coexpression", fontsize = 20)
plt.xlabel("Number of studied species with orthogroup", fontsize = 20)
sns.despine()

ax.grid(False)
ax.spines["bottom"].set_color("black")
ax.spines["left"].set_color("black")
sns.move_legend(ax, 'lower right')
#plt.xlim((-7,7))
# plt.ylim((0, 1250))
#plt.title("Conservation of orthogroup coexpression captures conserved  \n gene relationships missed in genewise comparisons", fontsize = 20)
plt.xticks(fontsize = 15,)
plt.yticks(fontsize = 15)
plt.savefig('/data/passala/Plots_for_projects/Cross_species_go_groups/Orthogroup_improvement_over_pairwise/orthogroup_improvement_over_pairwise.svg', transparent = True)
```
