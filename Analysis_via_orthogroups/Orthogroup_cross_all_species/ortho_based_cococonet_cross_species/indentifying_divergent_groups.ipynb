```python
import pandas as pd 
import GO_Enrichment_Calculator
import Go_annotations
```

```python
all_fc_scores = pd.read_csv('/data/passala/Generated_Tables/Comparing_all_orthogorups_across_plants/Ortho_fc_scores_full_table/avg_of_one_species_v_all.csv', index_col= 0 )
all_fc_scores
```

```python
all_fc_scores_sufficient_groups = all_fc_scores.dropna(thresh= 4)
```

```python
all_fc_scores_sufficient_groups
```

```python
all_fc_scores_sufficient_groups_high_mean = all_fc_scores_sufficient_groups.loc[all_fc_scores_sufficient_groups.mean(axis =1)>.8]
all_fc_scores_sufficient_groups_high_mean
```

```python
one_messed_up = all_fc_scores_sufficient_groups_high_mean.loc[all_fc_scores_sufficient_groups_high_mean.min(axis = 1)<.5]
```

```python
one_messed_up
```

```python
one_messed_up_with_arabi = one_messed_up.dropna(subset='arabidopsis')
one_messed_up_with_arabi
```

```python
ortho_to_gene = pd.read_csv('/data/passala/OrthoDB_data/NCBI_data/og_2_Genes_with_network_id.csv')
ortho_to_gene
```

```python
arabi_genes = ortho_to_gene.loc[ortho_to_gene['Orthogroup'].isin(one_messed_up_with_arabi.index) ]
arabi_genes
```

```python
arabi_genes_to_species = arabi_genes.loc[arabi_genes['Species'] == 3702 ]
arabi_genes_to_species
```

```python
list_of_arabidopsis_genes = arabi_genes_to_species['Symbol'].to_list()
```

```python
arabi_annotations  = Go_annotations.get_go_annotations_as_panda('arabidopsis')
```

```python
tables = GO_Enrichment_Calculator.generate_contigency_tables(go_annotations= arabi_annotations, gene_list= list_of_arabidopsis_genes)
```

```python
fishers_results = GO_Enrichment_Calculator.fishers_exact_on_contigency_tables(tables, arabi_annotations)
fishers_results
```

```python
fishers_results.sort_values(by = 'P_value').head(20)
```

```python
final_results = GO_Enrichment_Calculator.multi_hypothesis_correct_fishers_exact(fishers_results)
final_results
```
