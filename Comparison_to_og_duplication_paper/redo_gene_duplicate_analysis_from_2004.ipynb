```python
import CococoNet_reader
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ath_coconet = CococoNet_reader.read_cococonet('arabidopsis')
```

```python
duplication_guide = pd.read_csv('/data/passala/git/Plant_Go_Group_Analysis/Comparison_to_og_duplication_paper/gene_duplications_maps.csv')
duplication_guide
```

```python
ath_coconet
```

```python
ten_thousand_random = ath_coconet.melt(ignore_index= False).sample(10000)
```

```python
melted_coconet = ath_coconet.melt(ignore_index= False,var_name= 'gene2',value_name='coexpression').reset_index(drop= False, names = 'gene1')
melted_coconet
```

```python
ten_thousand_random
```

```python
present_in_cococonet = duplication_guide.loc[(duplication_guide['gene1'].isin(ath_coconet.index))&(duplication_guide['gene2'].isin(ath_coconet.index))]
```

```python
assigned_coexpression= present_in_cococonet.merge(right = melted_coconet, on = ['gene1','gene2'])
```

```python
ten_thousand_random['duplication age class'] = 'Random'
```

```python
ten_thousand_random_for_merge = ten_thousand_random[['value','duplication age class']]
ten_thousand_random_for_merge.reset_index(drop=True)
```

```python
ten_thousand_random_for_merge = ten_thousand_random_for_merge.rename(columns = {'value':'coexpression'})
```

```python
ten_thousand_random_for_merge
```

```python
assigned_coexpression_for_merge = assigned_coexpression[['coexpression','duplication age class']]
```

```python
assigned_coexpression_for_merge
```

```python
full_gene_plotter = pd.concat([assigned_coexpression_for_merge,ten_thousand_random_for_merge])
full_gene_plotter = full_gene_plotter.reset_index(drop=True)
```

```python
full_gene_plotter
```

```python
sns.histplot(data = full_gene_plotter,x = 'coexpression', hue='duplication age class',stat ='probability',common_norm=False, element = 'poly',fill = False)
```

```python
fig, ax = plt.subplots(figsize = (12,8))
fig.patch.set_facecolor('white')
ax = sns.histplot(data = full_gene_plotter,x = 'coexpression', hue='duplication age class',stat ='probability',common_norm=False, element = 'poly',fill = False,linewidth=3,palette=['tab:red','tab:blue','black'], bins = 30)
plt.ylabel("Frequency", fontsize = 20)
plt.xlabel("Normalised Coexpression", fontsize = 20)
sns.despine()

ax.grid(False)
ax.spines["bottom"].set_color("black")
ax.spines["left"].set_color("black")

ax.spines["bottom"].set_linewidth(3)
ax.spines["left"].set_linewidth(3)

#plt.xlim((-7,7))
# plt.ylim((0, 1250))
#plt.title("Current Threshold is 31", fontsize = 20)
plt.xticks(fontsize = 15,fontweight = 'regular')
plt.yticks(fontsize = 15,fontweight = 'regular')
plt.tick_params(width = 3,length = 6)
```

```python
fig, ax = plt.subplots(figsize = (12,8))
fig.patch.set_facecolor('white')
ax = sns.histplot(data = full_gene_plotter,x = 'coexpression', hue='duplication age class', element= 'step',stat ='probability',common_norm=False,linewidth=3,palette=['tab:red','tab:blue','black'], bins = 50)
plt.ylabel("Frequency", fontsize = 20)
plt.xlabel("Normalised Coexpression", fontsize = 20)
sns.despine()

ax.grid(False)
ax.spines["bottom"].set_color("black")
ax.spines["left"].set_color("black")

ax.spines["bottom"].set_linewidth(3)
ax.spines["left"].set_linewidth(3)
ax.legend(loc = 'upper left')
#plt.xlim((-7,7))
# plt.ylim((0, 1250))
#plt.title("Current Threshold is 31", fontsize = 20)
plt.xticks(fontsize = 15,fontweight = 'regular')
plt.yticks(fontsize = 15,fontweight = 'regular')
plt.tick_params(width = 3,length = 6)
```
