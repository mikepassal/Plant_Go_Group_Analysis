```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
arabi_genes = pd.read_csv('/data/passala/Module_paper_data/Medicago_grape_Redo_salt_pinot/medi_genes_de_results.csv', index_col= 0)
arabi_ortho = pd.read_csv('/data/passala/Module_paper_data/Medicago_grape_Redo_salt_pinot/medi_ortho_de_results_1_families_removed.csv', index_col= 0)
tomato_genes = pd.read_csv('/data/passala/Module_paper_data/Medicago_grape_Redo_salt_pinot/grape_genes_de_results.csv', index_col= 0)
tomato_ortho = pd.read_csv('/data/passala/Module_paper_data/Medicago_grape_Redo_salt_pinot/grape_ortho_de_results_1_families_removed.csv', index_col= 0)
```

```python
log_fc_change_evaluation_value = .9
log_fc_change_evaluation_value_negative = -1*log_fc_change_evaluation_value
```

```python
tomato_ortho
```

```python
trimmed_tomato_ortho = tomato_ortho.loc[tomato_ortho.index.isin(arabi_ortho.index)]
trimmed_arabi_ortho = arabi_ortho.loc[arabi_ortho.index.isin(trimmed_tomato_ortho.index)]
```

```python
sns.scatterplot(x =trimmed_tomato_ortho['log2FoldChange'], y = trimmed_arabi_ortho['log2FoldChange'])
plt.axvline(0, color="grey", linestyle="--")
plt.axhline(0, color="grey", linestyle="--")
```

```python
sns.scatterplot(x =trimmed_tomato_ortho['log2FoldChange'], y = trimmed_arabi_ortho['log2FoldChange'])
plt.axvline(log_fc_change_evaluation_value, color="grey", linestyle="--")
plt.axhline(log_fc_change_evaluation_value, color="grey", linestyle="--")
plt.axvline(log_fc_change_evaluation_value_negative, color="grey", linestyle="--")
plt.axhline(log_fc_change_evaluation_value_negative, color="grey", linestyle="--")
```

```python
merged_result = pd.DataFrame(data = zip(trimmed_tomato_ortho['log2FoldChange'], trimmed_arabi_ortho['log2FoldChange']), columns = ['Grape','Medicago'])

all_over_1 = merged_result.loc[(merged_result['Grape']>log_fc_change_evaluation_value) & (merged_result['Medicago']>log_fc_change_evaluation_value)]
all_under_1 = merged_result.loc[(merged_result['Grape']<log_fc_change_evaluation_value_negative) & (merged_result['Medicago']<log_fc_change_evaluation_value_negative)]
discord_bottom = merged_result.loc[(merged_result['Grape']>log_fc_change_evaluation_value) & (merged_result['Medicago']<log_fc_change_evaluation_value_negative)]
discord_top = merged_result.loc[(merged_result['Grape']<log_fc_change_evaluation_value_negative) & (merged_result['Medicago']>log_fc_change_evaluation_value)]
```

```python
discord_ratio = (len(all_over_1)+len(all_under_1))/(len(discord_bottom)+len(discord_top))
discord_ratio
```

```python
merged_result['Classification'] = 'Non-substantial'
merged_result['Classification'].loc[(merged_result['Grape']>log_fc_change_evaluation_value) & (merged_result['Medicago']>log_fc_change_evaluation_value)] = 'Substantial'
merged_result['Classification'].loc[(merged_result['Grape']<log_fc_change_evaluation_value_negative) & (merged_result['Medicago']<log_fc_change_evaluation_value_negative)]= 'Substantial'
merged_result['Classification'].loc[(merged_result['Grape']>log_fc_change_evaluation_value) & (merged_result['Medicago']<log_fc_change_evaluation_value_negative)]= 'Substantial'
merged_result['Classification'].loc[(merged_result['Grape']<log_fc_change_evaluation_value_negative) & (merged_result['Medicago']>log_fc_change_evaluation_value)]= 'Substantial'
```

```python
fig, ax = plt.subplots(figsize = (12,8))
fig.patch.set_facecolor('white')
sns.scatterplot(data = merged_result.loc[merged_result['Classification'] == "Substantial"] , x = 'Grape', y = 'Medicago')
plt.axvline(0, color="k", linestyle="--")
plt.axhline(0, color="k", linestyle="--")
plt.axvline(1, color="grey", linestyle="--")
plt.axhline(1, color="grey", linestyle="--")
plt.axvline(-1, color="grey", linestyle="--")
plt.axhline(-1, color="grey", linestyle="--")
plt.xlabel("Orthogroup-wise log2-fold change in Grape", fontsize = 20)
plt.ylabel("Orthogroup-wise log2-fold chagne in Medicago", fontsize = 20)
plt.title('Orthogroupwise Grape vs Medicago (Note Log2FC cutoff set to 1)\n 1.63 Concordant Groups per Discordant Group', fontsize = 23)
plt.xticks(fontsize = 18,)
plt.yticks(fontsize = 18)
```
