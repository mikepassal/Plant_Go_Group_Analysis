```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
arabi_genes = pd.read_csv('/data/passala/Module_paper_data/Sorghum_rice_heat_stress/sorg_genes_de_results.csv', index_col= 0)
arabi_ortho = pd.read_csv('/data/passala/Module_paper_data/Sorghum_rice_heat_stress/sorg_ortho_de_results_1_genefamily_removed.csv', index_col= 0)
tomato_genes = pd.read_csv('/data/passala/Module_paper_data/Sorghum_rice_heat_stress/rice_genes_de_results.csv', index_col= 0)
tomato_ortho = pd.read_csv('/data/passala/Module_paper_data/Sorghum_rice_heat_stress/rice_ortho_de_results_1_genefamily_removed.csv', index_col= 0)
```

```python
log_fc_change_evaluation_value = 1
log_fc_change_evaluation_value_negative = -1*log_fc_change_evaluation_value
```

```python
tomato_ortho
```

```python
trimmed_tomato_ortho = tomato_ortho.loc[tomato_ortho.index.isin(arabi_ortho.index)]
trimmed_arabi_ortho = arabi_ortho.loc[arabi_ortho.index.isin(trimmed_tomato_ortho.index)]
```

```python
fig, ax = plt.subplots(figsize = (12,8))
fig.patch.set_facecolor('white')
sns.scatterplot(x =trimmed_tomato_ortho['log2FoldChange'], y = trimmed_arabi_ortho['log2FoldChange'])
plt.axvline(0, color="k", linestyle="--")
plt.axhline(0, color="k", linestyle="--")
plt.axvline(1, color="grey", linestyle="--")
plt.axhline(1, color="grey", linestyle="--")
plt.axvline(-1, color="grey", linestyle="--")
plt.axhline(-1, color="grey", linestyle="--")
plt.xlabel("Orthogroup-wise log2-fold change in rice", fontsize = 20)
plt.ylabel("Orthogroup-wise log2-fold chagne in sorghum", fontsize = 20)
plt.xticks(fontsize = 18,)
plt.yticks(fontsize = 18)
```

```python
stats.spearmanr(a=trimmed_tomato_ortho['log2FoldChange'], b = trimmed_arabi_ortho['log2FoldChange'] )
```

```python
merged_result = pd.DataFrame(data = zip(trimmed_tomato_ortho['log2FoldChange'], trimmed_arabi_ortho['log2FoldChange']), columns = ['Rice','Sorghum'])

all_over_1 = merged_result.loc[(merged_result['Rice']>log_fc_change_evaluation_value) & (merged_result['Sorghum']>log_fc_change_evaluation_value)]
all_under_1 = merged_result.loc[(merged_result['Rice']<log_fc_change_evaluation_value_negative) & (merged_result['Sorghum']<log_fc_change_evaluation_value_negative)]
discord_bottom = merged_result.loc[(merged_result['Rice']>log_fc_change_evaluation_value) & (merged_result['Sorghum']<log_fc_change_evaluation_value_negative)]
discord_top = merged_result.loc[(merged_result['Rice']<log_fc_change_evaluation_value_negative) & (merged_result['Sorghum']>log_fc_change_evaluation_value)]
```

```python
discord_ratio = (len(all_over_1)+len(all_under_1))/(len(discord_bottom)+len(discord_top))
discord_ratio
```

```python
merged_result['Classification'] = 'Non-substantial'
merged_result['Classification'].loc[(merged_result['Rice']>log_fc_change_evaluation_value) & (merged_result['Sorghum']>log_fc_change_evaluation_value)] = 'Substantial'
merged_result['Classification'].loc[(merged_result['Rice']<log_fc_change_evaluation_value_negative) & (merged_result['Sorghum']<log_fc_change_evaluation_value_negative)]= 'Substantial'
merged_result['Classification'].loc[(merged_result['Rice']>log_fc_change_evaluation_value) & (merged_result['Sorghum']<log_fc_change_evaluation_value_negative)]= 'Substantial'
merged_result['Classification'].loc[(merged_result['Rice']<log_fc_change_evaluation_value_negative) & (merged_result['Sorghum']>log_fc_change_evaluation_value)]= 'Substantial'
```

```python
fig, ax = plt.subplots(figsize = (12,8))
fig.patch.set_facecolor('white')
sns.scatterplot(data = merged_result.loc[merged_result['Classification'] == "Substantial"] , x = 'Rice', y = 'Sorghum')
plt.axvline(0, color="k", linestyle="--")
plt.axhline(0, color="k", linestyle="--")
plt.axvline(1, color="grey", linestyle="--")
plt.axhline(1, color="grey", linestyle="--")
plt.axvline(-1, color="grey", linestyle="--")
plt.axhline(-1, color="grey", linestyle="--")
plt.xlabel("Orthogroup-wise log2-fold change in rice", fontsize = 20)
plt.ylabel("Orthogroup-wise log2-fold chagne in sorghum", fontsize = 20)
plt.title('Orthogroupwise Rice vs Sorghum (Note Log2FC cutoff set to 1)\n 6 Concordant Groups per Discordant Group', fontsize = 23)
plt.xticks(fontsize = 18,)
plt.yticks(fontsize = 18)
```
