```python
medi_metadata = pd.DataFrame(
    columns=["Temp"],
    data=["Hot", "Hot", "Hot", "Cold", "Cold", "Cold"],
    index=[
        "Tom Hot Rep 1",
        "Tom Hot Rep 2",
        "Tom Hot Rep 3",
        "Tom Cold Rep 1",
        "Tom Cold Rep 2",
        "Tom Cold Rep 3",
    ],
)
```

```python
genes_to_keep_medi = combined_medi.columns[combined_medi.sum(axis=0) >= 100]
medi_csv = combined_medi[genes_to_keep_medi]
inference_medi = DefaultInference(n_cpus=30)
dds_medi = DeseqDataSet(
    counts=medi_csv,
    metadata=medi_metadata,
    design_factors="Temp",
    refit_cooks=True,
    inference=inference_medi,
)
dds_medi.deseq2()
stat_res_medi = DeseqStats(dds_medi, inference=inference_medi)
stat_res_medi.summary()
```

```python
genes_to_keep = combined_grape.columns[combined_grape.sum(axis=0) >= 100]
combined_grape = combined_grape[genes_to_keep]
inference = DefaultInference(n_cpus=30)
dds = DeseqDataSet(
    counts=combined_grape,
    metadata=grape_metadata,
    design_factors="Temp",
    refit_cooks=True,
    inference=inference,
)
dds.deseq2()
stat_res = DeseqStats(dds, inference=inference)
stat_res.summary()
```

```python
full_results_grape = stat_res.results_df
full_results_grape
full_medi_results = stat_res_medi.results_df
```

```python
combined_grape = combined_grape.T
combined_medi = combined_medi.T
```

```python
full_results_grape.to_csv(
    "/data/passala/Module_paper_data/Medicago_grape_Redo_salt_pinot/grape_genes_de_results.csv"
)
full_medi_results.to_csv(
    "/data/passala/Module_paper_data/Medicago_grape_Redo_salt_pinot/medi_genes_de_results.csv"
)
```

```python
plt.scatter(
    x=full_medi_results["log2FoldChange"],
    y=full_medi_results["padj"].apply(lambda x: -np.log10(x)),
    s=1,
    label="Not significant",
)

# highlight down- or up- regulated genes
down_arabi = full_medi_results[
    (full_medi_results["log2FoldChange"] <= -3) & (full_medi_results["padj"] <= 0.0001)
]
up_arabi = full_medi_results[
    (full_medi_results["log2FoldChange"] >= 3) & (full_medi_results["padj"] <= 0.0001)
]

plt.scatter(
    x=down_arabi["log2FoldChange"],
    y=down_arabi["padj"].apply(lambda x: -np.log10(x)),
    s=3,
    label="Down-regulated",
    color="blue",
)
plt.scatter(
    x=up_arabi["log2FoldChange"],
    y=up_arabi["padj"].apply(lambda x: -np.log10(x)),
    s=3,
    label="Up-regulated",
    color="red",
)
plt.xlabel("logFC")
plt.ylabel("-logFDR")
plt.axvline(-3, color="grey", linestyle="--")
plt.axvline(3, color="grey", linestyle="--")
plt.axhline(2, color="grey", linestyle="--")
plt.legend()
```
