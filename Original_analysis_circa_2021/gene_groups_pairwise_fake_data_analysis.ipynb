```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fake_egad_results = pd.read_csv('/data/passala/Generated_Tables/Comparing_all_go_groups_across_species/fake_egad_results.csv',index_col= 0)
fake_egad_results =fake_egad_results[['grape AUC','arabidopsis AUC','mustard AUC','soybean AUC','medicago AUC','tomato AUC','potato AUC','rice AUC','brome AUC','sorghum AUC','maize AUC']]
fake_egad_results
```

```python
fake_go_terms_fc = pd.read_csv('/data/passala/Generated_Tables/Comparing_all_go_groups_across_species/fake_go_term_fc_results.csv',index_col = 0)
fake_go_terms_fc
```

```python
avg_fake_fc_and_fake_egad = pd.DataFrame(index = fake_go_terms_fc.index)
avg_fake_fc_and_fake_egad['AVG EGAD'] = fake_egad_results.mean(axis = 1)
avg_fake_fc_and_fake_egad['AVG FC'] = fake_go_terms_fc.mean(axis = 1)
avg_fake_fc_and_fake_egad
```

```python
fig, ax = plt.subplots(figsize = (15,15))
sns.scatterplot(data = avg_fake_fc_and_fake_egad, x = 'AVG EGAD',y = 'AVG FC', s = 30)
plt.xlabel("Average EGAD Scores of Fake Groups",fontsize = 20)
plt.ylabel("Average FC Scores of Fake Groups",fontsize = 20)
plt.title("Fake Groups, generated with an average group size of 16, do not show increase in FC variance with increasing EGAD Score",fontsize = 15)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
```
