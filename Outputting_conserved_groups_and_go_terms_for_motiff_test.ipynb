```python
group_table['Annotation'].unique()
```

```python
conserved_ribosome_and_chloroplast = group_table.loc[(group_table['Annotation'].isin(['Photosynthesis','Ribosomal']))&(group_table['Avg FC']>.85)&(group_table['Avg EGAD']>.85)]
conserved_ribosome_and_chloroplast
```

```python
coconet_plants = ['grape','arabidopsis','soybean','medicago','tomato','potato','rice','brome','sorghum','maize']
```

```python
current_species_annotations = Go_annotations.get_go_annotations_as_panda('grape')
current_species_annotations
```

```python
melted_annotations = current_species_annotations.melt(ignore_index=False)
```

```python
melted_annotations['value'].unique()
```

```python
conserved_go_group = conserved_ribosome_and_chloroplast.index
conserved_go_group
```

```python
melted_annotations.loc[(melted_annotations['GO_term'].isin(conserved_go_group))&(melted_annotations['value'] == 1)]
```

```python
conserved_ribosome_and_chloroplast
```

```python
master_list_of_all_genes = pd.DataFrame(columns = ['Loci','Go Number','Species','Name'])
conserved_go_group = conserved_ribosome_and_chloroplast.index
for species in coconet_plants:
    print(species)
    current_species_annotations = Go_annotations.get_go_annotations_as_panda(species)
    melted_annotations = current_species_annotations.melt(ignore_index=False)
    in_species_genes = melted_annotations.loc[(melted_annotations['GO_term'].isin(conserved_go_group))&(melted_annotations['value'] == 1)]
    built_dataframe = pd.DataFrame(columns = ['Loci','Go Number'], data = zip(in_species_genes.index, in_species_genes['GO_term']))
    built_dataframe['Species'] = species
    full_built_dataframe = built_dataframe.merge(right = conserved_ribosome_and_chloroplast[['Name']], how = 'left', right_index= True, left_on = 'Go Number' )
    master_list_of_all_genes = pd.concat([master_list_of_all_genes,full_built_dataframe])
    
```

```python
tomato_all_genes = master_list_of_all_genes.loc[master_list_of_all_genes['Species'] == 'tomato']
tomato_all_genes 
```

```python
tomato_converter = pd.read_csv('/data/CoCoCoNet/geneInfo/tomato_info.csv')
tomato_converter
tomato_all_genes = tomato_all_genes.merge(how = 'inner',right = tomato_converter[['EnsemblID','GeneSymbol']],left_on='Loci',right_on='GeneSymbol')
```

```python
gene_locations_tomato = pd.read_csv('/data/passala/Collaborator_Data/Ensembl_data_for_Go_group_analysis/tomato_all_gene_locations.txt', sep = '\t')
gene_locations_tomato
```

```python
labeled_genes = tomato_all_genes.merge(how = 'left', right = gene_locations_tomato [['Gene stable ID','Gene start (bp)','Gene end (bp)','Strand','Chromosome/scaffold name']],right_on = 'Gene stable ID',left_on = 'EnsemblID')
```

```python
labeled_genes['Regulatory Region Start'] = np.nan 
labeled_genes['Regulatory Region End'] = np.nan 
```

```python
labeled_genes['Regulatory Region Start'].loc[labeled_genes['Strand'] == 1] = labeled_genes.loc[labeled_genes['Strand'] == 1]['Gene start (bp)'] - 500
labeled_genes['Regulatory Region Start'].loc[labeled_genes['Strand'] == -1] = labeled_genes.loc[labeled_genes['Strand'] == -1]['Gene end (bp)']

labeled_genes
```

```python
labeled_genes['Regulatory Region End'].loc[labeled_genes['Strand'] == 1] = labeled_genes.loc[labeled_genes['Strand'] == 1]['Gene start (bp)']
labeled_genes['Regulatory Region End'].loc[labeled_genes['Strand'] == -1] = labeled_genes.loc[labeled_genes['Strand'] == -1]['Gene end (bp)']+500
labeled_genes
```
